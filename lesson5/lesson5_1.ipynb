```python
### 這是一個判斷區塊
### if 成立就跳出判斷
### if 不成立就跳到下一個elif
### 都不成立就跳到else

scores=int (input("輸入分數:"))
if scores <60: #比較嚴格的條件放最上面
    print("丁")

elif scores <70:
    print("丙")

elif scores <80:
    print("乙")

elif scores <90:
    print("甲")

else:
    print("優")
```

```python
"""兩個條件
-巢狀判斷
-邏輯運算子(and 或 or)
"""
# 國文 和 數學100--->1000元
# 國文 或 數學100--->500元
# 其他 ---> 0元

chiness=int(input("國文"))
math=int(input("數學"))

if chiness==100 and math ==100:
    print("$1000")
elif chiness==100 or math ==100:
    print("$500")
else:
    print("$0")
```

```python
"""
迴圈(loop)
for ... in ... :

"""

"""
phyon內建函數 range
class range(stop)
range (5) 5是引數值argument
串列值有5個元素

class range(start, stop, step=1)
step=1 是defaule value 可以不用寫
range(1,5) 有5-1=4個元素
"""

#說明range
for i in range(5): #5個元素
    print(i)

print ("="*30)

for j in range(1,10): #10-1=9個元素
    print(j)

print ("="*30)

#1~10加總

n=int(input("輸入n:"))
print(f"輸入n={n}")

total=0
for i in range(1,n+1):
    total +=i

print(f"total={total}")
```

```python
"""
巢狀迴圈

"""

for i in range(1,10):
    print(f"i={i}")

    for j in range(1,10):
        print(f"j={j}") #預設是end='\n'是連續列印
        #print(f"j={j}",end='') #end='空白'是換行列印
    print() #空一行列印

#print(*objects, sep=' ', end='\n', file=None, flush=False)


```

```python
#99乘法表
for i in range(1,10):
    for j in range(1,10):
        print(f"{i}*{j}={i*j}",end='\t') #end='\t'是tab
        
    print() #空一行列印
```
